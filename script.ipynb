```python
#  LLM Execution Optimizer for Edge AI – Colab Notebook
# Benchmarks PyTorch vs ONNX vs Quantized across multiple HuggingFace LLMs

#  Install requirements
!pip install -q transformers onnx onnxruntime psutil matplotlib

#  Imports
import torch
import time
import csv
import os
import psutil
import onnx
import onnxruntime as ort
import matplotlib.pyplot as plt
from transformers import AutoTokenizer, AutoModelForCausalLM
from collections import defaultdict

#  Model list
models_to_test = [
    "distilgpt2",
    "EleutherAI/gpt-neo-125M",
    "openai-community/gpt2"
]

prompt = "Explain 5G technology in simple terms."

#  Create output directory
os.makedirs("onnx_models", exist_ok=True)
os.makedirs("plots", exist_ok=True)

#  Profiler
def profile_inference(func):
    process = psutil.Process()
    mem_before = process.memory_info().rss / (1024 ** 2)
    start = time.time()
    _ = func()
    end = time.time()
    mem_after = process.memory_info().rss / (1024 ** 2)
    return round(end - start, 4), round(mem_after - mem_before, 2)

#  Benchmark function
results = []

for model_name in models_to_test:
    print(f"\n Benchmarking model: {model_name}")
    tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
    model = AutoModelForCausalLM.from_pretrained(model_name).to("cpu")
    model.eval()

    # PyTorch
    def torch_infer():
        inputs = tokenizer(prompt, return_tensors="pt").to("cpu")
        return model.generate(**inputs, max_length=100)

    latency_pt, mem_pt = profile_inference(torch_infer)
    print(f"PyTorch → Latency: {latency_pt}s, Memory: {mem_pt} MB")
    results.append([model_name, "PyTorch", latency_pt, mem_pt])

    # ONNX export
    try:
        inputs = tokenizer(prompt, return_tensors="pt")
        onnx_path = f"onnx_models/{model_name.replace('/', '_')}.onnx"

        torch.onnx.export(
            model,
            (inputs["input_ids"]),
            onnx_path,
            input_names=["input_ids"],
            output_names=["logits"],
            dynamic_axes={"input_ids": {0: "batch_size", 1: "sequence"}},
            opset_version=14
        )

        # ONNX inference
        def onnx_infer():
            session = ort.InferenceSession(onnx_path, providers=["CPUExecutionProvider"])
            return session.run(None, {"input_ids": inputs["input_ids"].numpy()})

        latency_onnx, mem_onnx = profile_inference(onnx_infer)
        print(f"ONNX → Latency: {latency_onnx}s, Memory: {mem_onnx} MB")
        results.append([model_name, "ONNX", latency_onnx, mem_onnx])
    except Exception as e:
        print(f" ONNX failed: {e}")
        results.append([model_name, "ONNX", "Error", "Error"])

    # Quantization
    try:
        quant_model = torch.quantization.quantize_dynamic(model, {torch.nn.Linear}, dtype=torch.qint8)

        def quant_infer():
            inputs = tokenizer(prompt, return_tensors="pt").to("cpu")
            return quant_model.generate(**inputs, max_length=100)

        latency_q, mem_q = profile_inference(quant_infer)
        print(f"Quantized → Latency: {latency_q}s, Memory: {mem_q} MB")
        results.append([model_name, "Quantized", latency_q, mem_q])
    except Exception as e:
        print(f" Quantization failed: {e}")
        results.append([model_name, "Quantized", "Error", "Error"])

#  Save results to CSV
with open("results.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Model", "Type", "Latency (s)", "Memory (MB)"])
    writer.writerows(results)

print("\n Results saved to results.csv")

#  Plotting
def load_results(csv_path="results.csv"):
    data = defaultdict(dict)
    with open(csv_path, mode="r") as file:
        reader = csv.DictReader(file)
        for row in reader:
            model = row["Model"]
            typ = row["Type"]
            try:
                latency = float(row["Latency (s)"]) if row["Latency (s)"] not in ["N/A", "Error"] else None
                memory = float(row["Memory (MB)"]) if row["Memory (MB)"] not in ["N/A", "Error"] else None
                data[model][typ] = {"latency": latency, "memory": memory}
            except ValueError:
                continue
    return data

def plot_metric(data, metric):
    labels = ["PyTorch", "ONNX", "Quantized"]
    models = list(data.keys())
    x = range(len(models))
    width = 0.2

    fig, ax = plt.subplots(figsize=(12, 6))
    colors = {"PyTorch": "skyblue", "ONNX": "orange", "Quantized": "green"}

    for i, typ in enumerate(labels):
        values = [data[m].get(typ, {}).get(metric) for m in models]
        values = [v if v is not None else 0 for v in values]
        ax.bar(
            [pos + (i - 1) * width for pos in x],
            values,
            width,
            label=typ,
            color=colors[typ],
        )

    ax.set_xticks(x)
    ax.set_xticklabels(models, rotation=15)
    ax.set_ylabel("Latency (s)" if metric == "latency" else "Memory (MB)")
    ax.set_title(f"Model Comparison – {metric.capitalize()}")
    ax.legend()
    plt.tight_layout()
    plt.savefig(f"plots/{metric}_plot.png")
    plt.show()

#  Load data and plot
data = load_results("results.csv")
plot_metric(data, "latency")
plot_metric(data, "memory")

#  Download helpers for GitHub upload
from google.colab import files
files.download("results.csv")
files.download("plots/latency_plot.png")
files.download("plots/memory_plot.png")
```
